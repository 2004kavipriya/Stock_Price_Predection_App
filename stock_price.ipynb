```python
import yfinance as yf
from datetime import datetime
import matplotlib.pyplot as plt

# Fetch Google stock data
end = datetime.now()
start = datetime(end.year - 20, end.month, end.day)
google_data = yf.download('GOOG', start=start, end=end)

# Check available columns
print("Available columns:", google_data.columns)

# Determine which column to use for calculations
if 'Adj Close' in google_data.columns:
    ma_column = 'Adj Close'
elif 'Close' in google_data.columns:
    ma_column = 'Close'
else:
    ma_column = None

# Calculate the 250-day moving average
if ma_column:
    google_data['MA_for_250_days'] = google_data[ma_column].rolling(window=250).mean()
    print(f"250-day moving average added using the '{ma_column}' column!")
else:
    print("Neither 'Adj Close' nor 'Close' columns are available in the data.")

# Plot the data
if ma_column:
    plt.figure(figsize=(15, 5))
    google_data[ma_column].plot(label=ma_column)
    google_data['MA_for_250_days'].plot(label='250-Day Moving Average')
    plt.xlabel("Years")
    plt.ylabel("Price")
    plt.title(f"Google Stock Price and 250-Day Moving Average ({ma_column})")
    plt.legend()
    plt.show()
else:
    print("Unable to plot data because the required columns are missing.")
```

```python
# Check for 'Adj Close', fallback to 'Close' if not present
if 'Adj Close' in google_data.columns:
    google_data['percentage_change_cp'] = google_data['Adj Close'].pct_change()
elif 'Close' in google_data.columns:
    google_data['percentage_change_cp'] = google_data['Close'].pct_change()
else:
    print("Neither 'Adj Close' nor 'Close' is available in the dataset.")
```

```python
from sklearn.preprocessing import MinMaxScaler

# Assuming close_price is a pandas Series
scaler = MinMaxScaler(feature_range=(0, 1))

# Reshape the data to 2D, as MinMaxScaler requires 2D data
scaled_data = scaler.fit_transform(close_price.values.reshape(-1, 1))

# Display the scaled data
print(scaled_data)
```

```python
len(scaled_data)
```

```python
x_data= []
y_data = []

for i in range(100, len(scaled_data)):
    x_data.append(scaled_data[i-100:i])
    y_data.append(scaled_data[i])

import numpy as np
x_data, y_data = np.array(x_data), np.array(y_data)

```

```python
splitting_len = int(len(x_data)*0.7)
x_train = x_data[:splitting_len]
y_train = y_data[:splitting_len]

x_test = x_data[splitting_len:]
y_test = y_train[splitting_len:]
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(x_test.shape)
```

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM
```

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense, Input

model = Sequential()

# Use an Input layer to define the input shape
model.add(Input(shape=(x_train.shape[1], 1)))

# Now, you can add your LSTM layers
model.add(LSTM(128, return_sequences=True))
model.add(LSTM(64, return_sequences=False))
model.add(Dense(25))
model.add(Dense(1))
```

```python
model.compile(optimizer='adam', loss = 'mean_squared_error')
```

```python
model.fit(x_train, y_train, batch_size=1, epochs = 2)
```

```python
model.summary()
```

```python
import matplotlib.pyplot as plt

def plot_graph(figsize, data, title):
    """
    Plots the given data with the specified title and figure size.

    Parameters:
        figsize (tuple): The size of the figure (width, height).
        data (DataFrame): DataFrame containing columns to plot.
        title (str): Title of the plot.
    """
    plt.figure(figsize=figsize)
    plt.plot(data['original_test_data'], label='Original Test Data', color='blue')
    plt.plot(data['predictions'], label='Predictions', color='orange')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Values')
    plt.legend()
    plt.show()

# Call the plot_graph function
plot_graph((15, 6), ploting_data, 'Test Data vs Predictions')
```
